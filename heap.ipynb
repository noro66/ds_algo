```python
# Min Heap (Heapify)
# Time: O(n), space: O(1)

A = [-4, 3, 1, 0, 2, 5, 10, 8, 12, 9]

import heapq
heapq.heapify(A)

A
```

```python
# Heap Pop (Extract min value)
# Time O(log n)

heapq.heappop(A)
```

```python
# heap Sort
# Time: O(n log n), Space O(n)
# NOTE: O(1) Space is possible via swaping, but this is complex

def heapsort(arr):
    heapq.heapify(arr)
    n = len(arr)
    new_list = [0] * n 
    for i in range(n):
        minn = heapq.heappop(arr)
        new_list[i] = minn
    print(new_list)

l = [9, 3, 5, 6, 1, 2, 4, 8, 7]

heapsort(l)
```

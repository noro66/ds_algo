```python
# Bubble Sort
# Time: O(n^2)
# Space: O(1)
def swap(lst, a, b):
    lst[a], lst[b] = lst[b], lst[a]

A = [1, 5, 3, 6, 4, 9, 0, 7, 8, 2]
ln = len(A)
# for i in range(ln):
#     for j in range(1, ln - i): 
#         if A[j] < A[j - 1]:
#             swap(A, j, j - 1)
A
```

```python
# Insertion Sort
# Time: O(n^2)
# space: O(1)

for i in range(1, ln):
    tmp = A[i]
    j = i - 1
    while tmp < A[j] and j >= 0:
        A[j + 1] =  A[j]
        j -= 1
    A[j + 1] = tmp
A
```

```python
# Selection Sort
# Time: O(n^2)
# Space: O(1)

minn = A[0]
for i in range(ln):
    minn = A[i]
    for j in range(i + 1, ln):
        if A[j] <  minn:
            minn, A[j] = A[j], minn
    A[i] = minn
A
```

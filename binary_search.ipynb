```python
A = [-3, -1, 0, 1, 4, 7]
# Naive O(n) searching
if 8 in A:
    print(True)
```

```python
# Traditional Binary Search
# Time: O(log n)
# Spane: O(1)
def binary_search(arr, target):
    N = len(arr)
    L = 0
    R = N - 1

    while L <= R:
        # M = (L + R ) // 2
        M = L + ((R - L) // 2)
        if arr[M] == target:
            return True
        elif target < arr[M]:
            R = M - 1
        else:
            L = M + 1

    return False
binary_search(A, 0)
```

```python
# Based on a condition
# [F, F, F, F, T, T, T, T]
B = [False, False, False, False, False, True, True, True, True, True,]
def binary_search_condition(arr):
    N = len(arr)
    L = 0
    R = N - 1
    while L < R:
        M = (L + R) // 2 
        if arr[M]:
            R = M
        else:
            L = M + 1
    return L
binary_search_condition(B)
```

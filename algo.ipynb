```python
import re

def verify_card_number(number):
    try:
        number = re.split(r'[ -]', number)
        number = "".join(number)
        num = int(number)
        is_second = False
        result = 0
        while num:
            n = num % 10
            num //= 10
            if is_second:
                n = n * 2
            is_second = not is_second
            if n != n % 10:
                first_num = n % 10
                n = 1 + first_num
            result += n
        if result % 10 == 0:
            return 'VALID!'
        return 'INVALID!'
    except ValueError:
        return 'INVALID!'

verify_card_number('4111-1111-1111-1111')
```
